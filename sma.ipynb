```python
tokennizer = tiktoken.encoding_for_model("gpt-4o")
print(tokennizer.name)
```

```python
tokennizer = tiktoken.get_encoding("o200k_base")
```

```python
promt="vous etes un assistant dans le domaine de l'analyse des sentiments"
```

```python
tokens=tokennizer.encode(promt)
```

```python
print(tokens)
```

```python
print(len(tokens))
```

```python
for token in tokens:
    t=tokennizer.decode_single_token_bytes(token)
    print(t)
```

```python
for token in tokens:
    t=tokennizer.decode_single_token_bytes(token).decode('utf-8')
    print(t)
```

```python
for token in tokens:
    t=tokennizer.decode_single_token_bytes(token).decode('utf-8')
    print(t, end="|")
```

```python
def tokens_count(prompt:str, model:str="o200k_base"):
    tokennizer = tiktoken.get_encoding(model)
    return len(tokennizer.encode(prompt))
```

```python
print(tokens_count("vous etes un assistant"))
```

```python
from langchain_openai import ChatOpenAI
```

```python
llm = ChatOpenAI(model="gpt-4o", temperature=0)
```

```python
resp = llm.invoke(input=[
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "c'est quoi un agent ai"}
])
```

```python
from IPython.display import Markdown
```

```python
print(resp.content)
```

```python
print(display(Markdown(resp.content)))
```
